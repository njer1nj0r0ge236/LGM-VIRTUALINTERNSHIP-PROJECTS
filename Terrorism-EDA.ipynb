## Attacks by country

```python
# Group data by country and count the number of attacks by country.

attacks_by_country = df_terrosism.groupby('country_txt').agg({'iyear':'count'}).reset_index()
```

```python
# Find country with highest number of attacks

country_most_attacks = attacks_by_country.loc[attacks_by_country['iyear'].idxmax()]
print(country_most_attacks)
```

```python
# Find the top 15 countries and the number of attacks

top_15_countries = df_terrosism.country_txt.value_counts()[:15]
print(top_15_countries)
```

```python
# Sort number of attacks in descending order

attacks_by_country.sort_values(by = 'iyear', ascending = False)
```

```python
# Plot A bar for the top 15 countries and the number of attacks

plt.subplots(figsize=(16,6))
plt.bar(top_15_countries.index,top_15_countries.values)
plt.title('Top 15 Countries ')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
# Create a pivot table of the number of terrorist attacks in each country

df_terrosism_pivot = df_terrosism.pivot_table(index='country_txt', values='iyear',aggfunc='count')
```

```python
# Plot a heatmap for the pivot table

sns.heatmap(df_terrosism_pivot, cmap='Blues')
```

```python
# Bar chart for the number of attacks by country

plt.bar(attacks_by_country.index, attacks_by_country)
plt.xlabel('country')
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.show()
```

## Attacks per city

```python
# Group by city and count the number of cities per attack

city_attacks = df_terrosism.groupby('city').size()
```

```python
# Sort the data in descending order

city_attacks = city_attacks.sort_values(ascending=False)
```

```python
# Pie chart for number of attacks per city

plt.pie(city_attacks.head(10),
        labels = city_attacks.head(10).index)
plt.title('Top 10 cities with most attacks')
plt.show()
```

## Attacks per region

```python
# Group per region and count the number of attacks

region_attacks = df_terrosism.groupby('region_txt').size()
```

```python
# Region latitudes and longitudes
longitude = df_terrosism.columns('longitude')
latitude = df_terrosism.columns('latitude')
 
```

```python
# Scatter plot chart for number of attacks per region

plt.scatter(df_terrosism.longitude.values(),
            df_terrosism.latitude.values(),
            s=region_attacks,
            alpha=0.5)
plt.title('Number of attacks per region')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

## Attacks per year

```python
# Convert the date column to a datetime object

df_terrosism['iyear'] = pd.to_datetime(df_terrosism['iyear'],format='%Y')
```

```python
# Group by year and count number of attacks per year

year_attacks = df_terrosism.groupby('iyear').size()
```

```python
# Histogram for the number of attacks per year

plt.hist(year_attacks,
         bins=range(1970,2018,1),
         edgecolor='red')
plt.title('Histogram for Number of attacks per year')
plt.xlabel('Year')
plt.ylabel('Number of attacks')
plt.show()
```

## Attacks per month

```python
# Convert date column to datetime object

df_terrosism['date'] = pd.to_datetime(df_terrosism['iyear'].astype(str) + df_terrosism['imonth'].astype(str) + df_terrosism['iday'].astype(str) , format='%Y,%m,%day')
```

```python
# Group by date and count number of attacks per month

month_attacks = df_terrosism.groupby('date').size()
```

```python
# Line chart for the number of attacks per month

plt.plot(month_attacks.index, month_attacks.values)
plt.title('Number of attacks per month')
plt.xlabel('Date')
plt.ylabel('Number of attacks')
plt.show()
```

## Type of most attacks 

```python
# Group by attack type and count the number of attacks for each type

type_attacks = df_terrosism.groupby('attacktype1_txt').size()
```

```python
# Find which attack type has most attacks

most_type_attacks = type_attacks.idxmax()
print(f'The attack type with most attacks is {most_type_attacks}.')
```

```python
# Bar chart for number of attacks by type

plt.bar(type_attacks.index, type_attacks.values)
plt.title('Bar Chart for Number of Attacks by Type')
plt.xlabel('Attack Type')
plt.ylabel('Number of attacks')
plt.show()
```
